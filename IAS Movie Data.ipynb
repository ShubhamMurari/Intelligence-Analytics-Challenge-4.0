### Create a fucntion to check the % of Null Values in each column:

```python
def nulls(df):
    #fig,ax=plt.figsize(18,12)
    nulls = pd.DataFrame(round(df.isnull().sum()/len(df.index)*100,2),columns=['null_percent'])
    #sns.barplot(x='index',y='null_percent',data=nulls.reset_index())
    return nulls[nulls['null_percent']!=0.00].sort_values('null_percent',ascending=False)
```

#### 1. Imputaion of Years-

##### Logic is we will check for top 5 values of Months with maximum number of movies and then randomly impute the values. 

```python
#Following are the top 5 years in which maximum movies got released: 

df_vol1['Movie Release Month'].value_counts().head(5)
```

##### Random Imputation of years by using 'np.random.choice'

```python
df_vol1['Movie Release Month'] = df_vol1['Movie Release Month'].fillna(np.mean(df_vol1['Movie Release Month']))
```

```python
df_vol1['Movie Release Month'] = np.round(df_vol1['Movie Release Month'], decimals=0)
```

```python
df_vol1['Movie Release Month'] = df_vol1['Movie Release Month'].astype(int)
```

## Train-Test split and Model Fitting: 

```python
X = df_nonzero.drop('Box Office Gross',axis=1)
y = df_nonzero['Box Office Gross']
```

### Split the dataset into Test and Train:

```python
from sklearn.model_selection import train_test_split

X_train_org, X_test_org, y_train, y_test = train_test_split(X, y, random_state = 0)
```

### Scale the Dataset:

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train_org)
X_test = scaler.fit_transform(X_test_org)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Using Random Forest Regression: 

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 500, random_state = 42)

#Train Model on training data < X_train >
rf_reg = rf.fit(X_train, y_train)
```

```python
print(rf_reg.score(X_train, y_train)*100)
```

```python
print(rf_reg.score(X_test, y_test)*100)
```

### Get the R Squared value after Random Forest:

```python
from sklearn.metrics import r2_score, mean_squared_error

r2_score(y_test, rf_reg.predict(X_test))
```

#### Around 63%

### Get the RMSE after Random Forest

```python
np.sqrt(mean_squared_error(y_test, rf_reg.predict(X_test)))
```

#### RMSE after Random Forest is coming around 1.9 and our y_test mean is 15.5

```python
y_test.mean()
```

## Trying xgBoost now: 

```python
from xgboost import XGBRegressor

rfx = XGBRegressor(n_estimators = 1500, random_state = 42)

rfx_reg = rfx.fit(X_train, y_train)
```

```python
print(rfx_reg.score(X_train, y_train)*100)
```

```python
print(rfx_reg.score(X_test, y_test)*100)
```

### Get the R Squared value after xgBoost:

#### R-Squared value is coming around 68%
